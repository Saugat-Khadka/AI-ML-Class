```python
import pandas as pd
import numpy as np


house_price = [245,312,279,308,199,219,405,324,319,255]

# Generate 10 random square foot values between 1000 and 5000
square_foot = [1400,1600,1700,1875,1100,1550,2350,2450,1425,1700]

# Create a DataFrame
house_data = pd.DataFrame({
    'house_price': house_price,
    'square_foot': square_foot
})

print(house_data)
```

```python
sns.scatterplot(house_data, x="square_foot", y="house_price")
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()

X = house_data[["square_foot"]]
y = house_data["house_price"]

model=reg.fit(X, y)

print(reg.coef_)
print(reg.intercept_)
```

```python
model.predict(X=[[1250]])
```

```python
sns.scatterplot(house_data, x="square_foot", y="house_price")
plt.plot(X, reg.predict(X), color='red')
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

accuracy = r2_score(y,model.predict(X))
mse = mean_squared_error(y,model.predict(X))
mae = mean_absolute_error(y,model.predict(X))

print(" R2 accuracy is: "+str(accuracy))
print("mean squared error is: "+str(mse))
print("mean absolute error is: "+str(mae))
```

# DAY 5
multiple and polynomial regression

```python
import pandas as pd
import numpy as np
house_price = [245,312,279,308,199,219,405,324,319,255]

# Generate 10 random square foot values between 1000 and 5000
square_foot = [1400,1600,1700,1875,1100,1550,2350,2450,1425,1700]

no_of_rooms = [8,10,11,13,6,9,15,16,9,10]

# Create a DataFrame
df = pd.DataFrame({
    'house_price': house_price,
    'square_foot': square_foot,
    'no_of_rooms':no_of_rooms
})
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()

X = df[["square_foot","no_of_rooms"]]
y = df["house_price"]

model = reg.fit(X, y)

print(model.coef_) # slope
print(model.intercept_) # c
```

```python
predicted_value = model.predict(X=[[1600, 3]])
print(predicted_value[0])
```
