```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
path = r"C:\Users\ksaug\OneDrive\Desktop\Day\Day 10\Data\Mall_Customers.csv"
df = pd.read_csv(path)
```

```python
df.info()
```

```python
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})
df.drop(columns=["CustomerID"], inplace=True)
df.head(5)
```

```python
K = 5
KMeans_model = KMeans(
    n_clusters=K,
    init="k-means++",
    n_init=10,
    max_iter=300,
    random_state=42
)
KMeans_model.fit(df)
```

```python
KMeans_model.cluster_centers_
```

```python
KMeans_model.labels_
```

```python
df["Clusters"] = KMeans_model.labels_
df.tail(10)
```

```python
cluster_average = df.groupby("Clusters").mean()
cluster_average
```

```python
new_customers = pd.DataFrame([
    {"Gender": 0, "Age": 28, "Annual Income (k$)": 70, "Spending Score (1-100)": 50},
    {"Gender": 1, "Age": 40, "Annual Income (k$)": 120, "Spending Score (1-100)": 20},
    {"Gender": 0, "Age": 23, "Annual Income (k$)":203, "Spending Score (1-100)": 79},
])

predictions = KMeans_model.predict(new_customers)
predictions
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c=df["Clusters"])
centroids = KMeans_model.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], c='red', marker='*')
```
