```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
#print(iris.data)
print(iris.feature_names) # gives the names of the features in the dataset :column_names
print(iris.target_names) # gives the number of classes in the dataset
print(iris.data.shape) #gives the number of rows and columns in the dataset
```

```python
x = iris.data
y = iris.target
# y = iris.target_names[iris.target] # gives the target names for the given target values
```

```python
model = DecisionTreeClassifier().fit(x, y)
```

```python
predictions = model.predict([[1.1, 1.2, 3.0, 2.3]])
# 1.1 is the sepal length and 1.2 is the sepal width
# 3.0 is the petal length and 2.3 is the petal width

print(iris.target_names[predictions[0]]) # gives the predicted class for the given input features
```

```python
from sklearn.ensemble import RandomForestClassifier

model_2 = RandomForestClassifier().fit(x, y)

prediction_2 = model_2.predict([[1.1, 1.2, 3.0, 2.3]])

print(iris.target_names[prediction_2[0]]) # gives the predicted class for the given input features
```
