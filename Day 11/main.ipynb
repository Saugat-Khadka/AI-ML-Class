```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X = np.array([[500], [1000], [1500], [2000], [2500], [3000], [3500]], dtype=float)
y = np.array([[50], [100], [150], [200], [250], [300], [350]], dtype=float) #in lakhs
```

```python
# Normalize data
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)
```

```python
#Split train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
print(scaler_X.inverse_transform((X_test)))
```

```python
#Neural Network model

model = Sequential([
  Input(shape=(1,)),
  Dense(10, activation='relu'),
  Dense(5, activation='relu'),
  Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=20)
```

```python
print("X-test",scaler_X.inverse_transform(X_test))

predicted = scaler_y.inverse_transform(model.predict(X_test))
print("predicted Price",predicted.flatten())
```

```python
## classification using neural network
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelBinarizer
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#load dataset
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1,1)
```

```python
#onehot encoding
encoder = LabelBinarizer()
y_onehot = encoder.fit_transform(y)
```

```python
#normalize features
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
```

```python
# split train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_onehot, test_size=0.2, random_state=42)
```

```python
# neural network
model = Sequential([
    Dense(10, input_shape=(4,), activation='relu'),
    Dense(8, activation='relu'),
    Dense(3, activation='softmax')
])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
#train
model.fit(X_train, y_train, epochs=100, verbose=0)
```

```python
# evaluate
loss, accuracy = model.evaluate(X_test, y_test, verbose = 0)
print(f"Test loss: {loss:.4f}")
print(f"Test accuracy: {accuracy:.4f}")
new_data = np.array([[5.1, 3.5, 1.4, 0.2]])
new_data_scaled = scaler_X.transform(new_data)
prediction = model.predict(new_data_scaled)
```

```python
predicted_class_decoded = encoder.inverse_transform(prediction)
print(f"Predicted class: {predicted_class_decoded[0]}")
```
